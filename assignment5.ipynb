```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron

# Define the input features and target labels
x1 = [1, 0, 0, 1]
x2 = [1, 0, 1, 0]
x = [[1, 1], [0, 0], [0, 1], [1, 0]]
y = [1, 0, 0, 0]

# Initialize and train the Perceptron model
clf = Perceptron(max_iter=100).fit(x, y)
print("Classes of the model:", clf.classes_)
print("Intercept of the decision boundary:", clf.intercept_)
print("Coefficients of the decision boundary:", clf.coef_)

# Calculate the decision boundary line (y = mx + c)
ymin, ymax = -1, 2
w = clf.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(ymin, ymax)
yy = a * xx - (clf.intercept_[0]) / w[1]

# Plot the data points and the decision boundary
plt.figure(figsize=(4, 4))
ax = plt.axes()
ax.scatter(x1, x2, c=y, cmap='bwr', edgecolor='k')
plt.plot(xx, yy, 'k-')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
plt.title('Perceptron Decision Region')
plt.show()
```
